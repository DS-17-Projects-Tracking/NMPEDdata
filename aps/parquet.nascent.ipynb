## Read in parquet file


```python
import pandas as pd
import requests
import zipfile
import io
```

```python
url = 'https://github.com/DS-17-Projects-Tracking/NMPEDdata/raw/refs/heads/main/aps/Perseverance_Survey_132_cols.csv.zip'

response = requests.get(url)
with zipfile.ZipFile(io.BytesIO(response.content)) as zip_ref:
  csv_filename = [x for x in zip_ref.namelist() if x.endswith(".csv")][0]
  with zip_ref.open(csv_filename) as csvfile:
    df = pd.read_csv(csvfile, encoding='utf-16', delimiter='\t')
df.shape
```

```python
def try_numeric(data_series):
  '''
  The answer to pd.to_numeric( coerce="error" )

  Usage:
    df.apply(try_numeric)
  '''
  data_types = ["int", "float", "datetime64[ns]", "category", "str"]
  for dtype in data_types:
    if ( dtype == "category" and data_series.nunique(dropna=False) > 50 ) :
      continue
    try:
      return data_series.astype(dtype)
    except:
      pass
  return data_series
```
