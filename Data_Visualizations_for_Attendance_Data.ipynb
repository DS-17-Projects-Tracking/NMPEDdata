### üìà Line: State trends for other school districts with over 10,000 students

```python
df
```

```python
noncharter_filter = df[(df['SchNumb'] > 900) & (df['SchNumb'] < 500000 )]
```

```python
state_es_filter= (df['SchNumb'].isin( list(school_dir_df[school_dir_df['Level'] == 'es']['ID']))) & (df['Subgroup'] == 'All') & noncharter_filter
df[state_es_filter]
state_ms_filter= (df['SchNumb'].isin( list(school_dir_df[school_dir_df['Level'] == 'ms']['ID']))) & (df['Subgroup'] == 'All')& noncharter_filter
df[state_ms_filter]
state_hs_filter= (df['SchNumb'].isin( list(school_dir_df[school_dir_df['Level'] == 'hs']['ID']))) & (df['Subgroup'] == 'All')& noncharter_filter
df[state_hs_filter]

midtolarge_dists = list(df[(df['SchNumb']<100) & (df['Subgroup'] == 'All') & (df['Number_Attending']>10000)]['SchNumb'].unique())
state_dist_list = list(school_dir_df['District'][school_dir_df['ID'].isin(midtolarge_dists)]) ##Pop of district over 10000

## High school state wide averages by distrist to graph
district_trends_hs_df= pd.DataFrame()
for i in state_dist_list:
  dist_filter = df['SchNumb'].isin(list(school_dir_df[school_dir_df['District'] == i]['ID']))
  for year in range(2020,2025):
    district_trends_hs_df = pd.concat([district_trends_hs_df,pd.Series([i, year, df[state_hs_filter & dist_filter & (df['Year'] == year)]['Regular_Attendance_Rate'].mean()])], axis = 1)

district_trends_hs_df = district_trends_hs_df.transpose().reset_index(drop = True)
district_trends_hs_df.columns = ['District', 'Year','Regular_Attendance_Rate']


## Middle school state wide averages by distrist to graph
district_trends_ms_df= pd.DataFrame()
for i in state_dist_list:
  dist_filter = df['SchNumb'].isin(list(school_dir_df[school_dir_df['District'] == i]['ID']))
  for year in range(2020,2025):
    district_trends_ms_df = pd.concat([district_trends_ms_df,pd.Series([i, year, df[state_ms_filter & dist_filter & (df['Year'] == year)]['Regular_Attendance_Rate'].mean()])], axis = 1)

district_trends_ms_df = district_trends_ms_df.transpose().reset_index(drop = True)
district_trends_ms_df.columns = ['District', 'Year','Regular_Attendance_Rate']
district_trends_ms_df

## Elementary school state wide averages by distrist to graph
district_trends_es_df= pd.DataFrame()
for i in state_dist_list:
  dist_filter = df['SchNumb'].isin(list(school_dir_df[school_dir_df['District'] == i]['ID']))
  for year in range(2020,2025):
    district_trends_es_df = pd.concat([district_trends_es_df,pd.Series([i, year, df[state_es_filter & dist_filter & (df['Year'] == year)]['Regular_Attendance_Rate'].mean()])], axis = 1)

district_trends_es_df = district_trends_es_df.transpose().reset_index(drop = True)
district_trends_es_df.columns = ['District', 'Year','Regular_Attendance_Rate']
district_trends_es_df
```

```python
state_fig_es = px.line(
    district_trends_es_df,
    x = 'Year',
    y= 'Regular_Attendance_Rate',
    # facet_col= 'Cluster',
    color = 'District',
    title = 'Average Elementary School Attendance by District Statewide<br><sup>Districts with over 10000 students'
    )
state_fig_es.update_layout(
    xaxis=dict(
        range=[2020, 2024],  # Set x-axis range from 2020 to 2024
        dtick=1,      # Set x-axis tick interval to 1
        ),
    yaxis=dict(
        range=[0, 1],  # Set y-axis range from 0 to 1
        dtick=0.1,
        ),
    width=600,
    height=400
)

state_fig_es.show()
state_fig_es.to_html('AveElemAttendancebyDist.html')

state_fig_ms = px.line(
    district_trends_ms_df,
    x = 'Year',
    y= 'Regular_Attendance_Rate',
    # facet_col= 'Cluster',
    color = 'District',
    title = 'Average Middle School Attendance by District Statewide<br><sup>Districts with over 10000 students'
    )
state_fig_ms.update_layout(
    xaxis=dict(
        range=[2020, 2024],  # Set x-axis range from 2020 to 2024
        dtick=1,      # Set x-axis tick interval to 1
        ),
    yaxis=dict(
        range=[0, 1],  # Set y-axis range from 0 to 1
        dtick=0.1,
        ),
    width=600,
    height=400
)

state_fig_ms.show()
state_fig_es.to_html('AveMidAttendancebyDist.html')


state_fig_hs = px.line(
    district_trends_hs_df,
    x = 'Year',
    y= 'Regular_Attendance_Rate',
    range_x = (2020,2024),
    # facet_col= 'Cluster',
    color = 'District',
    title = 'Average High School Attendance by District Statewide: <br><sup>Districts with over 10000 students',
    )

state_fig_hs.update_layout(
    xaxis=dict(
        range=[2020, 2024],  # Set x-axis range from 2020 to 2024
        dtick=1,      # Set x-axis tick interval to 1
        ),
    yaxis=dict(
        range=[0, 1],  # Set y-axis range from 0 to 1
        dtick=0.1,
        ),
    width=600,
    height=400
)

state_fig_hs.show()
state_fig_es.to_html('AveHighAttendancebyDist.html')
```
